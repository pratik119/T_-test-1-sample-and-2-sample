```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
army_weight=stats.norm.rvs(loc=78,size=30,scale=1)
```

```python
army_weight
```

```python
sns.histplot(army_weight,kde=True,bins=10)
```

```python
np.mean(army_weight)
```

```python
navy_weight=stats.norm.rvs(loc=78,size=30,scale=1)
```

```python
sns.histplot(navy_weight,kde=True)
```

```python
np.mean(navy_weight)
```

```python
ttest_score,p_value=stats.ttest_ind(army_weight,navy_weight)
```

```python
p_value
```

```python
if p_value<=0.05:
    print("Null Hypothesis is rejected! There is a significant difference between mean of two samples")
else:
    print('Null Hypothesis is accected!There is not a significant difference between mean of two samples')
```
