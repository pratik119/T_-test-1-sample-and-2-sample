```python
data['ne']=data['Age'][data['Pclass'] == 1]  #vlookup function in Excel? age when pclass=1-- copy to-- ne
```

```python
data.head()
```

```python
help(sns.boxplot)
```

```python
def inage(lst):
    age=lst[0]
    pclass=lst[1]
    
    if pd.isnull(age):
        if pclass==1:
            return 39
        elif pclass==2:
            return 29
        else:
            return 21
    else:
        return age
```

```python
data['Age']=data[['Age','Pclass']].apply(inage,axis=1)
```

```python
data.head()
```

```python
sns.heatmap(pd.isnull(data),yticklabels=False)
```

```python
data.drop(['Cabin'],inplace=True,axis=1)
```

```python
sns.heatmap(data.isnull(),yticklabels=False)
```

```python
data['Male']=pd.get_dummies(data['Sex'],drop_first=True)
```

```python
data[['q','s']]=pd.get_dummies(data['Embarked'],drop_first=True)
```

```python
data.head()
```

```python
data.drop(['PassengerId','Lname','Name','Sex','Ticket','Fare','ne'], inplace=True,axis=1)
```

```python
data.head()
```

```python
data.drop(['Embarked'],inplace=True,axis=1)
```

```python
data.dropna(inplace=True)
```

```python
data.drop(['SibSp','Parch'],axis=1, inplace=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
help(train_test_split)
```

```python
x_train,x_test,y_train,y_test=train_test_split(data.drop(['Survived'],axis=1),data['Survived'],test_size=0.30,random_state=101)
```
